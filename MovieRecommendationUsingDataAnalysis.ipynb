```python
# so merge the ratings and movie dataset

df = movies_df.merge(ratings_df, on='movieId')
df
```

```python
# creating a title and stored it in list of recommended movies.
T_s = 'Toy Story (1995)'
recommended_movies = []

#find movie in dataset and sort it as per ratings
movie_db = df[df['title'] == T_s]\
            .sort_values(by='rating', ascending=False)

#get first 5 users who liked this movie
for user in movie_db.iloc[:5]['userId'].values:
    
    #get rated movies from this user
    rated_movies = df[df['userId'] == user]
    
    #get the five biggest rated movie by this user
    rated_movies = rated_movies[rated_movies['title'] != T_s]\
                        .sort_values(by='rating', ascending=False)\
                        .iloc[:5]
    
    #add these to the recommendations
    recommended_movies.extend(list(rated_movies['title'].values))
    
 
recommended_movies = np.unique(recommended_movies)
    
for movie in recommended_movies:
    print(movie)
```

```python
#now weigth each movie by the similarity on the genre features

gmovie_genres = df[df['title']==T_s].iloc[0]['genres'].split('|')
scores = {}

for movie in recommended_movies:
    movied = df[df['title']==movie].iloc[0]
    movie_genres = movied['genres'].split('|')
    score = 0
    
    #scoring on how many gmovie_genre can be found in movie_genres?
    for gmovie_genre in gmovie_genres:
        if gmovie_genre in movie_genres:
            score += 1
            
    scores[movie] = score
    
#Sort them on score and reserve it, because the bigger the score the better
recommended_movies = sorted(scores, key=lambda x: scores[x])[::-1]
```

```python
#the recommendations are now weighted

for movie in recommended_movies:
    print(movie)
```
